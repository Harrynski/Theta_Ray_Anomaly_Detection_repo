```python
import pandas as pd
import numpy as numpy
from sklearn.neighbors import LocalOutlierFactor
```

```python
df = pd.read_excel(rf'TXS que el collector no ve NOV23 al final del mes.xlsx')
```

```python
df = df.drop(columns='Unnamed: 0').set_index('CreatedAt')
```

```python
X = pd.get_dummies(df[['Client', 'VendorCode', 
       'CollectMethod', 'TargetCountry',  'NetAmountUSD',
       'SourceCountry']])
```

```python
from sklearn.ensemble import IsolationForest
clf = IsolationForest(random_state=0, contamination=0.1).fit(X)
#clf.predict([[0.1], [0], [90]])
```

```python
df['Anomaly_scores']  = clf.decision_function(X)
```

```python
df['is_anomaly']  = clf.predict(X)
```

```python
df.sort_values(by='is_anomaly')
```

```python
anomalias = df[df.is_anomaly==-1].sort_values(by='Anomaly_scores', ascending=True)
```
